```python
def value_history_scraper(av_serial_number):
    web_response = requests.get(fr"https://www.utahcounty.gov/landrecords/property.asp?av_serial={av_serial_number}", headers = headers)
    soup = bs4.BeautifulSoup(web_response.text)
    divs = soup.find_all('div', attrs = {'class' : "TabbedPanelsContent"})
    if len(divs) == 0:
        return
    Value_History = divs[1]
    all_table_rows = Value_History.find_all('tr')
    x = 0
    Years = []
    Real_Estate_Values_Tot = []
    Improvements_Tot = []
    for i in all_table_rows:
        if (x == 0) or (x == 1):
            x += 1
            continue
        all_tds = i.find_all('td')
        Years.append(all_tds[0].text)
        Real_Estate_Values_Tot.append(all_tds[4].text)
        Improvements_Tot.append(all_tds[8].text)
    data = {'Years': Years, 'Real_Estate_Values_Tot': Real_Estate_Values_Tot, 'Improvements_Tot': Improvements_Tot}
    value_history_df = pd.DataFrame(data)
    value_history_df['Real_Estate_Values_Tot'] = value_history_df['Real_Estate_Values_Tot'].str.strip('$').str.replace(',', '').astype(float)
    value_history_df['Improvements_Tot'] = value_history_df['Improvements_Tot'].str.strip('$').str.replace(',', '').astype(float)
    return value_history_df
```
